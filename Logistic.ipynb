```python
def step_gradient(x,y,lr,beta):
    M=len(beta)
    N=len(x)
    l_beta_slope=np.zeros(M)
    l_beta_slope.reshape(M,1)
    for j in range(M):
        t=0
        for i in range(N):
            z=np.dot(x[i],beta)
            hyp=float(1)/float((1+np.exp(-z)))
            t=t+(hyp-y[i])*x[i][j]
        t=t/N
        l_beta_slope[j]=t
    beta=beta-lr*l_beta_slope
    return beta
```

```python
def gd(x,y,lr,itr):
    M=len(x)
    beta=np.zeros(x.shape[1])
    beta.reshape(x.shape[1],1)
    for i in range(itr):
        b=step_gradient(x,y,lr,beta)
        beta=b
    return b
```
