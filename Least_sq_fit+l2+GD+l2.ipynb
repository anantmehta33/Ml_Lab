```python
def fit(X,y):
    x_new = np.column_stack((np.ones(len(X)),X))
    lamda=10
    identity=np.identity(len(x_new))
    #x_new-->3*3
    a=np.dot(x_new.T,x_new)#3*3
    a1=a+lamda*identity
    beta=np.dot(np.linalg.inv(a),np.dot(x_new.T,y))
    return beta
```

```python
def run():
    beta=fit(X,y)
    print(beta)
    #pred
    x=np.array([1,4,80])
    y_pred=np.dot(x,beta)
    print('y pred is ',y_pred)
```

```python
run()
```

```python
def cost(y_pred,y):
    u=(y-y_pred)**2
    v=(y-y.mean())**2
    return 1-u/v
```

```python
def step_gradient(x,y,lr,beta):
    M=len(beta)
    N=len(x)
    l_beta_slope=np.zeros(M)
    l_beta_slope.reshape(M,1)
    #print(l_beta_slope)
    #print(x[2][1])
    for j in range(M):
        t=0
        for i in range(N):
            t=t+(np.dot(x[i],beta)-y[i])*x[i][j]
        t=t/N
        #print(t)
        l_beta_slope[j]=t
        #print(l_beta_slope)
        #print(l_beta_slope.shape)
    #print(l_beta_slope)
    beta=beta-lr*l_beta_slope
    return beta
```

```python
def gd(x,y,lr,itr):
    M=len(x)
    beta=np.zeros(x.shape[1])
    #print(beta)
    beta.reshape(x.shape[1],1)
    for i in range(itr):
        b=step_gradient(x,y,lr,beta)
        beta=b
        #print(b)
        y_pred=np.dot(x,b)
        print(i, " Cost: ", cost(y_pred,y).sum())
    return b
```

```python
def run():
    X_new = np.column_stack((np.ones(len(X)),X))
    itr=1000
    lr=0.001
    beta=gd(X_new,y,lr,itr)
```
